```python
sample_mean=np.mean(sample_data)
sample_deviation=np.std(sample_data,ddof=1)
```

```python
sample_mean

```

```python
sample_deviation
```

```python
# H0= avg daily wages of urban worker =384 (as stated by International labour Organisation)
# H1=avg daily wages != 384

# two-tail test without population standard deviation (t-test)
alpha=0.05
#critical value
dof=len(sample_data)-1
t_critical_upper=s.t.ppf(1-(alpha/2),dof)
t_critical_lower=s.t.ppf(alpha/2,dof)
```

```python
#observed value
t_stats=(sample_mean-384)/(sample_deviation/np.sqrt(len(sample_data)))                           
                           
```

```python
if t_stats>=t_critical_lower and t_stats<=t_critical_upper:
    print('We fail to reject Null Hypothesis(H0)')
else:
    print('We reject Null Hypothesis')
```
